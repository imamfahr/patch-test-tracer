```python
import pandas as pd
import datetime as dt
```

```python
df1 = pd.read_excel('Summary_patch_test2.xlsx').set_index(['Unit','PM service']).drop(['Kidney Loop machine Type','Start KL Time','End of KL Time','Component','Oil Type'],axis=1)
df2 = pd.read_excel('Service history.xlsx').set_index(['Unit','PM service']).sort_index()
```

```python
df1.head(10)
```

```python
df2.head(50)
```

```python
df_combine = pd.merge(df1,df2,left_index=True,right_index=True)
df_combine[['Sample Date','Complete Date']]=df_combine[['Sample Date','Complete Date']].apply(pd.to_datetime)
df_combine.head(10)
```

```python
df_combine['days deviation'] = (df_combine['Sample Date']-df_combine['Complete Date']).dt.days
df_combine['Patch test at this PM'] = df_combine['days deviation'].apply(lambda x: 'True' if -24 <= x <= 24 else 'False')
df_combine.head(10)
```

```python
df_combine_2021 = df_combine.loc[(df_combine['Sample Date']>= '2021-01-01')].filter['']
df_combine_2021.head(10)
```

```python
df_combine_2021['Patch test at this PM'].value_counts()
```

```python
df_combine_2021_true = df_combine_2021[df_combine_2021['Patch test at this PM']=='True']
df_combine_2021_true.head()
```

```python
df_combine_all = df_combine[df_combine['Patch test at this PM']=='True'].drop_duplicates(subset=['Hour Meter when service'])
df_combine_all.head()
```
